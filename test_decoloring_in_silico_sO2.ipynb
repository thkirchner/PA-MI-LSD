```python
model_path = drive+"trained_models"
in_silico_data_set = "SET03_sO2_TRAIN"
START_WL_ID = 0

STOP_WL_ID = 15+1
SORT = False
MOD = '-unsorted-augmented'

wl_select_str = str(START_WL_ID)+"-"+str(STOP_WL_ID-1)

## load the "ground truth spectra" of the sulfate solutions
HB = pd.read_csv(
    "./sO2_ref_spectra.csv", header=0)
NM = np.arange(680,981,20)
HB = HB[HB["lambda/nm"].isin(NM)]

initial_guess_rCu = 0
initial_guess_tMix = 0
HBO2_SPECTRUM = np.asarray(HB["mua(HbO)//cmM"])
HBO2_SPECTRUM = HBO2_SPECTRUM[START_WL_ID:STOP_WL_ID]
HB_SPECTRUM = np.asarray(HB["mua(Hb)//cmM"])
HB_SPECTRUM = HB_SPECTRUM[START_WL_ID:STOP_WL_ID]

x0 = np.array([initial_guess_rCu, initial_guess_tMix], dtype=float)

def res_LU_sulfates(coeffs, wavelength_measured, spectrum_measured):
    # the linear unmixing residum 
    # coeffs[]: rCu, tMix
    spectrum_linear_combination = (coeffs[0]*coeffs[1] * HBO2_SPECTRUM +
                                   (coeffs[1] - coeffs[0]*coeffs[1]) * HB_SPECTRUM)
    return spectrum_measured - spectrum_linear_combination

## load the random forests trained on the in silico data set
rf_milsd = pickle.load(open(model_path+"/"+wl_select_str+MOD+'/rf_mi-lsd_'+in_silico_data_set+'_n100_d30', 'rb'))
rf_lsd = pickle.load(open(model_path+"/"+wl_select_str+MOD+'/rf_lsd_'+in_silico_data_set+'_n100_d30', 'rb'))
## load the ffnn (without dropout) trained on the in silico data set
nn_milsd = torch.load(model_path+"/"+wl_select_str+MOD+'/nn_mi-lsd_'+in_silico_data_set+'_b10000_e100_l0.01_d0.0')
nn_lsd = torch.load(model_path+"/"+wl_select_str+MOD+'/nn_lsd_'+in_silico_data_set+'_b10000_e100_l0.01_d0.0')
```

```python
## in silico test set
## the linear unmixing will take long (>5 min)
import time

set_name = "SET03_sO2_TEST"
i_path = drive+"data_in_silico/MI-LSD_"
o_path = drive+"trained_models/0-15-unsorted-augmented/"

y, X, S = preprocess_np_for_lsd(i_path+set_name, start_wl_id=0, stop_wl_id=16)
y_mi, X_mi = preprocess_np_for_milsd(i_path+set_name, start_wl_id=0, stop_wl_id=16, sort=SORT)

LU = np.zeros_like(y)
start_time = time.time()
for i, r in enumerate(y):
    if i%10000==0:
        print(str(i)+"/"+str(len(y)))
    LU[i] = least_squares(res_LU_sulfates, x0, args=(NM, X[i,:])).x[0]
print("unmixed "+str(y.size)+" samples")
print("%s seconds for LU" % (time.time() - start_time))
```

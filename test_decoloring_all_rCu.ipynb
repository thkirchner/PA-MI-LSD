```python
model_path = drive+"trained_models"
in_silico_data_set = "SET03_TRAIN"
START_WL_ID = 0

STOP_WL_ID = 15+1
SORT = False
MOD = '-unsorted-augmented'

wl_select_str = str(START_WL_ID)+"-"+str(STOP_WL_ID-1)

## load the "ground truth spectra" of the sulfate solutions
SULFATES = pd.read_csv(
    "./rCu_ref_spectra.csv", header=0)
NM = np.arange(680,981,20)
SULFATES = SULFATES[SULFATES["lambda/nm"].isin(NM)]
initial_guess_rCu = 0
initial_guess_tMix = 0
CUSO4_SPECTRUM = np.asarray(SULFATES["mua(Cu)//cm"])[::-1]
CUSO4_SPECTRUM = CUSO4_SPECTRUM[START_WL_ID:STOP_WL_ID]
NISO4_SPECTRUM = np.asarray(SULFATES["mua(Ni)//cm"])[::-1]
NISO4_SPECTRUM = NISO4_SPECTRUM[START_WL_ID:STOP_WL_ID]

x0 = np.array([initial_guess_rCu, initial_guess_tMix], dtype=float)

def res_LU_sulfates(coeffs, wavelength_measured, spectrum_measured):
    # the linear unmixing residum 
    # coeffs[]: rCu, tMix
    spectrum_linear_combination = (coeffs[0]*coeffs[1] * CUSO4_SPECTRUM +
                                   (coeffs[1] - coeffs[0]*coeffs[1]) * NISO4_SPECTRUM)
    return spectrum_measured - spectrum_linear_combination

## load the random forests trained on the in silico data set
rf_milsd = pickle.load(open(model_path+"/"+wl_select_str+MOD+'/rf_mi-lsd_'+in_silico_data_set+'_n100_d30', 'rb'))
rf_lsd = pickle.load(open(model_path+"/"+wl_select_str+MOD+'/rf_lsd_'+in_silico_data_set+'_n100_d30', 'rb'))
## load the ffnn (without dropout) trained on the in silico data set
nn_milsd = torch.load(model_path+"/"+wl_select_str+MOD+'/nn_mi-lsd_'+in_silico_data_set+'_b10000_e100_l0.01_d0.0')
nn_lsd = torch.load(model_path+"/"+wl_select_str+MOD+'/nn_lsd_'+in_silico_data_set+'_b10000_e100_l0.01_d0.0')
```
